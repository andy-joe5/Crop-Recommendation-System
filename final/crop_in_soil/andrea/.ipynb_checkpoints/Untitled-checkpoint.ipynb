```python
label = pd.Series([])
soil_type = pd.Series([])
```

```python
for i in range(len(df)):
    if df["     RAIN"][i] <= 400:
        label[i]="rice"
        soil_type[i]=1
    elif df["     RAIN"][i] <= 600:
        label[i]="cofee"
        soil_type[i]=2
    elif df["     RAIN"][i] == 1023:
        label[i]="onion"
        soil_type[i]=3
    elif df["     RAIN"][i] <= 800:
        label[i]="tomato"
        soil_type[i]=4
    else:
        label[i]= "potato"
        soil_type[i]=5
# inserting new column with values of list made above	
df.insert(7, "soil_type", soil_type)
df.insert(8, "label", label)
```
